```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/sample_data/housing.csv')


columns = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']


bins = [5, 10, 15, 20]


fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, col in enumerate(columns):
    data = df[col].dropna()


    counts, bin_edges = np.histogram(data, bins=bins, density=True)


    axes[i].bar(bin_edges[:-1], counts, width=np.diff(bin_edges), edgecolor='black')
    axes[i].set_title(f'Density of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Density')


plt.tight_layout()
plt.show()
```

```python
import numpy as np
from numpy import sqrt,log,pi,exp
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/sample_data/housing.csv')
data
def FindMLEAndLLv(x):
  m = data['latitude'].mean()
  v = data['latitude'].var()
  mle = (1/(sqrt(2*pi*v)))*(exp(-(x-m)**2/(2*v)))
  return mle , np.log(mle)
x_values = [50,75,80]
for x in x_values:
  mle,lle = FindMLEAndLLv(x)
  print(f"the max likelyhood value of {x} is : {mle}")
  print(f"the log-likelihood value of {x} is : {lle}")
  print("\n")
```
