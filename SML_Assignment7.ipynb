```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
salary = pd.read_csv('/content/Salary_Data.csv')
x=salary.drop('Salary',axis=1)
y=salary['Salary']
name='X0'
values=[1]*len(x)
x.insert(0, name, values)
x
"After inserting the column, x now contains two columns:"
"The intercept term is multiplied 1 by It allows us to calculate the bias term"
```

```python
# prompt: plt garph between X,Y for all features

import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
for feature in x.columns[1:]:  # Exclude the 'Target' column
  plt.figure(figsize=(8, 6))
  plt.scatter(x[feature], y)
  plt.xlabel(feature)
  plt.ylabel('Target')
  plt.title(f'Scatter Plot of {feature} vs Target')
  plt.show()
```

```python

x_train1,x_test1,y_train1,y_test1=train_test_split(x,y,test_size=0.2,random_state=10)

#25
x_train2,x_test2,y_train2,y_test2=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python

print(x_train1.shape)
print(x_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

```python

w=[1,10]
w=np.array(w)
print(w)
```

```python
def predict(X,w):
  return np.dot(X,w)

```

```python
def compute_error(Y,yp):
  mae=np.mean(np.abs(Y-yp))
  mse=np.mean((Y-yp)**2)
  return mae,mse
```

```python
"This code uses gradient descent to minimize the error between the predicted values and the actual values"
def linearRgression(X,w,Y,lr=0.001,it=100):
  for _ in range(it):
    yp=predict(X,w)
    myp=np.mean((Y-yp))
    for i in range(len(w)):
      mean=np.mean(X[:,i])
      w[i]=w[i]-lr*2*myp*(-mean)
    if(_+1)%100==0:
      print(w)
  return w
```

```python
print(w.copy())
w_trained_20 =linearRgression(x_train1.values,w.copy(),y_train1.values)
print()
print(w_trained_20)
```

```python
print(w.copy())
w_trained_25 = linearRgression(x_train2.values,w.copy(),y_train2.values)
print()
print(w_trained_25)
"Print the final trained weights (w_trained_20) to see how the weights have been optimized after training"
```

```python
"The code is performing prediction using the trained weights for two different test sets (20% and 25%)."
yp_20 = predict(x_test1.values, w_trained_20)
yp_25 = predict(x_test2.values, w_trained_25)
```

```python
"It calculates MAE (which gives the average error in absolute terms) and MSE (which gives the average squared error)."
mae_20, mse_20 = compute_error(y_test1.values, yp_20)
mae_25, mse_25 = compute_error(y_test2.values, yp_25)
print(f"Test Error (20%): MAE = {mae_20}, MSE = {mse_20}")
print(f"Test Error (25%): MAE = {mae_25}, MSE = {mse_25}")
```

```python
for col in x.columns[1:]:
  plt.figure(figsize=(8, 6))
  plt.scatter(x_test1[col], y_test1, color='blue', label='Actual Data')
  plt.plot(x_test1[col], yp_20, color='red', label='Best Fit Line')
  plt.xlabel(col)
  plt.ylabel('Target')
  plt.title(f"{col} vs Target with Best Fit Line")
  plt.legend()
  plt.show()
"ploting best with line with testing data 1"
```
