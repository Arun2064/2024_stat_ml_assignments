```python
import pandas as pd
file_path = "/content/diabetes_data_upload.csv"
df = pd.read_csv(file_path)
print(df)
```

```python
y=df['class']
print(y)

x=df.drop('class',axis=1)
print(x)
```

```python
mapping = {'Male ': 1, 'Female ': 0}
df['Gender'] = df['Gender'].map(mapping)
print(df)
```

```python
replace= df.replace({'No': 0, 'Yes': 1})
print(replace)
```

```python

replace= replace.replace({'Positive': 1, 'Negative': 0})
print(replace)
```

```python
import matplotlib.pyplot as plt

for column in x.columns:
    plt.figure()
    plt.scatter(replace[column].astype(str), y)
    plt.xlabel(column)
    plt.ylabel('Class')
    plt.title(f'Scatter plot of {column} vs Class')
    plt.show()
```
