```python
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

# Scatter Plot
plt.scatter(age, height)
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Scatter Plot of Age vs Height")
plt.show()

# Bar Chart
plt.bar(people, weight)
plt.xlabel("People")
plt.ylabel("Weight")
plt.title("Bar Chart of Weight for each Person")
plt.show()
```

```python
# Histogram of Age
plt.hist(age, bins=5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```
